```python
import pandas as pd
import numpy as np
import yfinance as yf
from fbprophet import Prophet
import matplotlib.pyplot as plt
from fbprophet.plot import plot_plotly

class FTahmin():
    def __init__(self,cur):
        self.cur=cur
    def tarih(self,start,end):
        para=yf.download(self.cur,start,end)
        self.para=para
        return self.para
    def grafik(self):
        para=self.para
        plt.plot(para.index,para['Close'])
        plt.show()
    def plotly(self):
        para=self.para
        plot_plotly(self.model,self.tahmin)
        plt.show()
        
    def tahmin(self,periyot,frekans):
        df=self.para.reset_index()
        df=df[["Date","Close"]]
        df.columns=["ds","y"]
        model=Prophet()
        model.fit(df)
        self.model=model
        future=model.make_future_dataframe(periods=periyot,freq=frekans)
        tahmin=model.predict(future)
        self.tahmin=tahmin
        model.plot(tahmin)
        plt.show()
        


aselsan=Ftahmin("asels.is")
aselsan.tarih("2010-01-01","2022-01-01")
aselsan.tahmin(360,"D")
```
